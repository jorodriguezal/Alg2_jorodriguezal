Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
target='AGTAGATGTAGTTAACTTTAATCTC'
a=secuence.find(target)#Se encuentra la posición de la primera aparición de target en secuence con la función de python
print(a)
```

```python
longitud = len(target) #Se define la longitud de target para ser consultada solo una vez
primer = target[0] #Se consulta el primer caracter de target
for i in range(len(secuence)): #Se itera sobre la longitud de toda la secuencia
    if primer == secuence[i]: #Si el primer caracter de target coincide con la posición i de la secuencia
        if secuence[i:i+longitud] == target: #Se parte secuence entre i e i+longitud de target y se compara si este string es igual a target 
            print('Posición inicial:', i,'\nPosición final:', i+longitud) #Se devuelve la primera y última posición de target en secuence
            break
```
